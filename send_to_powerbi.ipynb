```python
import requests
import json

# Power BI API Endpoint
POWER_BI_URL = "https://api.powerbi.com/v1.0/myorg/datasets/{dataset_id}/tables/{table_name}/rows"

# Authentication (Replace with your API key or OAuth token)
HEADERS = {
    "Content-Type": "application/json",
    "Authorization": "Bearer YOUR_ACCESS_TOKEN"
}

# Sample Data (Replace with actual results)
shap_results = {"feature1": 0.12, "feature2": -0.05, "feature3": 0.30}
lime_results = {"feature1": 0.10, "feature2": -0.02, "feature3": 0.25}
prediction_results = {"customer_id": 12345, "prediction": "Approved", "confidence": 0.95}

# Combine Results into JSON Format
data_to_send = {
    "shap": shap_results,
    "lime": lime_results,
    "prediction": prediction_results
}

# Send Data to Power BI
response = requests.post(POWER_BI_URL, headers=HEADERS, json=data_to_send)

# Handle Response
if response.status_code == 200:
    print("Data successfully sent to Power BI.")
else:
    print(f"Error sending data: {response.status_code} - {response.text}")
```
